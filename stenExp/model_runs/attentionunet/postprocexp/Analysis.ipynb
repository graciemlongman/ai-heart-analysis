# Post processing threshold analysis

```python
import pandas as pd
data = pd.read_csv('pp_scores.csv')
data = data.drop([2,4,6,8,10])
data.at[0,'Threshold'] = 'thresh_0'
data
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,5))

ax.bar(data['Threshold'], data['F1'], color='skyblue')

for i, v in enumerate(data['F1']):
    ax.text(i, v, f"{v:.3f}", ha='center', va='bottom', fontsize=9)

ax.grid(alpha=0.5)
ax.set(title='F1 score by post processing threshold for Attention UNet using Adam', xlabel='Threshold', ylabel='F1 score')
plt.ylim(min(data['F1'])-0.001, max(data['F1'])+0.001)
plt.show()
```
